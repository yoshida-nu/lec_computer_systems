```
num = [1, 2, 3, 4, 5, 6, 7]
for left in range(0, len(num) // 2):
	right = len(num) - left - 1
	tmp = num[right]
	num[right] = num[left]
	num[left] = tmp
#以下は問題にはない記述だがnumの内容を確認するために実行する）
print(num)
```
